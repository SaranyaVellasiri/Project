```python
df_country=pd.read_excel('Country-Code.xlsx')
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head(2)
```

```python
##To check Data Types
final_df.dtypes
```

```python
final_df.columns
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_names
```

```python
country_val=final_df.Country.value_counts().values
```

```python
country_val
```

```python
## Pie Chart- Top 3 countries that uses zomato
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

Observation:Zomato maximum records or transaction are from India After that USA and then United Kingdoms

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating _Count'})
```

```python
ratings
```

# Observation
1.When Rating is between 4.5 to 4.9---> Excellent
2.When Rating are between 4.0 to 3.4--->very good
3.when Rating is between 3.5 to 3.9----> good
4.when Rating is between 3.0 to 3.4----> average
5.when Rating is between 2.5 to 2.9----> average
6.when Rating is between 2.0 to 2.4----> Poor

```python
sns.barplot(x='Aggregate rating',y='Rating _Count',data=ratings)
```

```python
sns.barplot(x='Aggregate rating',y='Rating _Count',hue='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

# Observation:

Not Rated count is very high
Maximum number of rating are between 2.5 to 3.4

```python
sns.countplot(x="Rating color",data=ratings,palette=['blue','red','orange','yellow','green','green'])
```
